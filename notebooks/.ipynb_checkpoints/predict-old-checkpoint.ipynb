```python
from genome import Genome
genome = Genome('/Users/jpaggi/Downloads/MPRA/snps/hg19.fa')
```

```python
import numpy as np
bases = ['A', 'C', 'G', 'T']
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bps = {}
with open('bps_introns.tsv') as fp:
    for line in fp:
        chrom, pos, _, _, _, strand, _, start, end = line.split('\t')[:9]
        pos, start, end = int(pos), int(start), int(end)
        # No hit!
        if start == -1: continue
            
        # Check if reasonable
        if strand == '+':
            dist = end - pos - 1
        else:
            dist = pos - start
            
        if not 10 < dist < 50: continue
        
        # Add branchpoint
        key = (chrom, pos, strand)
        if key not in bps: bps[key] = []
        if key not in bps[key]:
            bps[key] += [(start, end)]
print len(bps)

plt.hist(map(lambda x: len(x), bps.values()), bins = 50)
plt.show()
```

```python
L = 55

def onehot(seq):
    X = np.zeros((len(seq), len(bases)))
    for i, char in enumerate(seq):
        X[i, bases.index(char)] = 1
    return X

X = []
y = []
for bp, introns in bps.items():
    start, end = introns[0]
    chrom, pos, strand = bp
    if strand == '+':
        begin, stop = end - L, end
        bp = L + pos - end
    else:
        begin, stop = start, start + L
        bp = L - pos + start - 1
    if not K < bp < L-K: continue
    seq = genome.get_seq(chrom, begin, stop, strand)
    X += [onehot(seq).T.reshape(1, -1, L)]
    
    _y = np.zeros((stop - begin,))
    _y[bp] = 1
    y += [_y]

X, y = np.vstack(X), np.vstack(y)
print X.shape, y.shape
```

```python
b = 40000
X_train, X_test = X[:b], X[b:]
y_train, y_test = y[:b], y[b:]
```

```python
def matching(preds, true):
    total, match = 0, 0
    for p, t in zip(preds, true):
        if np.argmax(p) == np.argmax(t):
            match += 1
        total += 1
    return match, total

def plot_examples(preds, true, k = 10):
    for p, t in zip(preds[:k], true[:k]):
        plt.plot(t)
        plt.plot(p)
        plt.show()
```

# PWM model

```python
K = 4
counts = np.zeros((2*K+1, 4))
total = 0
for target, seq in zip(y_train, X_train):
    bp = target.argmax()
    #print seq.T[bp-K: bp+K+1]
    #break
    #print bp
    counts = counts + seq.T[bp-K: bp+K+1]
    total += 1
print counts
pwm = counts / total
print pwm
```

```python
def pwm_pred(X, pwm):
    scores = []
    for seq in X:
        s = [0] * K
        for i in range(seq.shape[1] - K*2):
            s += [np.sum(seq.T[i:i+2*K+1] * pwm)]
        scores += [np.array(s)]
    return np.vstack(scores)

print matching(pwm_pred(X_test, pwm), y_test)
print matching(pwm_pred(X_train, pwm), y_train)
```

# Model trainer

```python
def train(model, X_train, X_test, y_train, y_test, epochs = 20):
    test, train = [], []
    for i in range(epochs):
        model.fit(X_train, y_train, epochs = 1, verbose = 0)
        print "{} of {} epochs".format(i, epochs)
        if i % 2 == 0:
            test += [matching(model.predict(X_test), y_test)[0]]
            train += [matching(model.predict(X_train), y_train)[0]]
    print len(test), len(train)
    print matching(model.predict(X_test), y_test)
    print matching(model.predict(X_train), y_train)
    plt.plot(range(0, epochs, 2),
             np.array(test) / float(y_test.shape[0]))
    plt.plot(range(0, epochs, 2),
             np.array(train) / float(y_train.shape[0]))
    return model
```

# Convolutional LSTM

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers.convolutional import Conv2D
from keras.layers.core import Reshape
```

```python
model = Sequential()
model.add(Conv2D(16, (1, 4), input_shape=(1, 4, L),
                 padding = 'same', activation = 'relu'))
model.add(Conv2D(16, (1, 4), input_shape=(1, 16, L),
                 padding = 'same', activation = 'relu'))
model.add(Conv2D(16, (1, 4), input_shape=(1, 16, L),
                 padding = 'same', activation = 'relu'))
model.add(Reshape((-1, 32)))
model.add(LSTM(L, input_shape=(L, 32)))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
train(model, X_train.reshape(-1, 1, 4, L), X_test.reshape(-1, 1, 4, L),
     y_train, y_test, 25)
```

```python
model = Sequential()
model.add(Conv2D(32, (1, 6), input_shape=(1, 4, L),
                 padding = 'same', activation = 'relu'))
model.add(Conv2D(32, (1, 6), input_shape=(1, 32, L),
                 padding = 'same', activation = 'relu'))
model.add(Reshape((-1, 32)))
model.add(LSTM(32, input_shape=(L, 32), return_sequences = True))
model.add(LSTM(L, input_shape=(L, 32)))
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
train(model, X_train.reshape(-1, 1, 4, L), X_test.reshape(-1, 1, 4, L),
     y_train, y_test, 50)
```
